```python
# area conversion into sq kms
def kanal_to_sq_kms(kanals):
  # slicing to strings to get integer numbers
  kanals = kanals[:-6].replace(',', '')
  kanals_float = float(kanals)
  return kanals_float/1976.843

def marlas_to_sq_kms(marlas):
  # slicing to strings to get integer numbers
  marlas = marlas[:-6].replace(',', '')
  marlas_float = float(marlas)
  return marlas_float/39536.8



df.loc[df['area'].str.endswith('Marla', na=False), 'area'] = df.loc[df['area'].str.endswith('Marla', na=False), 'area'].apply(marlas_to_sq_kms)
df.loc[df['area'].str.endswith('Kanal', na=False), 'area'] = df.loc[df['area'].str.endswith('Kanal', na=False), 'area'].apply(kanal_to_sq_kms)
df
```

```python
# finding outliers from price data


# room outliers
def find_outliers(column, interested_variable, interested_variable_column_name):
  (df['property_type']==interested_variable).mean()
  price = df.loc[(df[interested_variable_column_name] == interested_variable), column]
  Q1 = price.quantile(0.25)
  Q3 = price.quantile(0.75)

  IQR = Q3 - Q1
  upper_bound = Q1 - 1.5 * IQR
  lower_bound = Q3 + 1.5 * IQR

  min = price.quantile(0)
  max = price.quantile(1)
  return upper_bound, lower_bound, min, max

# printing outliers for all and min, max
print('.................', 'lower outlier,', 'Upper outlier,', 'min,', 'max,')
print('Room Outliers :',find_outliers('price', 'Room', 'property_type'))
print('House Outliers : ',find_outliers('price', 'House', 'property_type'))
print('Flat Outliers : ',find_outliers('price', 'Flat', 'property_type'))
print('Farm House Outliers : ',find_outliers('price', 'Farm House', 'property_type'))
print('Penthouse Outliers : ',find_outliers('price', 'Penthouse', 'property_type'))
print('Lower Portion Outliers : ',find_outliers('price', 'Lower Portion', 'property_type'))
print('Upper Portion Outliers : ',find_outliers('price', 'Upper Portion', 'property_type'))
```

```python
# removing rows with unexpected prices

# rooms
# < 8000 rent && < 15000 sale
# house
# < 21000 rent && 10 laks sale
# penthouse
# < 13000 rent && 18 laks sale 
# upper portion
# < 12000 rent && 20 laks sale 
# lower portion
# < 35000 rent && 20,laks  sale 

# farm house
# 90000 rent && 25 laks sale

#flat
# 25k rent && 50k sale
# df2 = df
# df.loc[df['property_type'] == 'Room', ['property_type', 'price']]
# df.loc[(df['property_type'] == 'Room'), 'property_type'] = 
# df.loc[(df['property_type'] == 'Room'), 'price'].mean()

def print_metadata(property_type):
    print(f'\n\n\n\n\n\n\n\n\ntotal {property_type} columns', (df['property_type'] == property_type).sum())

    # seperating property type
    property_column_seperated = df.loc[df['property_type'] == property_type, ['property_type', 'price', 'purpose']]
    print(property_column_seperated)
    
    # splitting into rent and sale
    property_column_sale = property_column_seperated.loc[property_column_seperated['purpose'] == 'For Sale', ['property_type', 'price']]
    property_column_rent = property_column_seperated.loc[property_column_seperated['purpose'] == 'For Rent', ['property_type', 'price']]


    # Sale details 
    print(f'{property_type} sale details')
    print('max', property_column_sale['price'].max())
    print('min', property_column_sale['price'].min())
    print('mean', property_column_sale['price'].mean())
    print('median', property_column_sale['price'].median())
    print('mode', property_column_sale['price'].mode())

    # Rent details 
    print(f'{property_type} Rent details')
    print('max', property_column_rent['price'].max())
    print('min', property_column_rent['price'].min())
    print('mean', property_column_rent['price'].mean())
    print('median', property_column_rent['price'].median())
    print('mode', property_column_rent['price'].mode())

# mean, median, mode values are too far from reality, unexpected

def drop_irrelevant_price_rows(min_bound_rent, min_bound_sale, max_bound_rent, max_bound_sale, property_type):
    # droping ones below minimum bound
    if min_bound_rent:
        df.drop(df[(df['price'] <  min_bound_rent) & (df['property_type'] == property_type) & (df['purpose'] == 'For Rent')].index, inplace=True)
    if min_bound_sale:
        df.drop(df[(df['price'] <  min_bound_sale) & (df['property_type'] == property_type) & (df['purpose'] == 'For Sale')].index, inplace=True)

    # droping ones above max bound
    if max_bound_rent:
        df.drop(df[(df['price'] >  max_bound_rent) & (df['property_type'] == property_type) & (df['purpose'] == 'For Rent')].index, inplace=True)
    if max_bound_sale:
        df.drop(df[(df['price'] >  max_bound_sale) & (df['property_type'] == property_type) & (df['purpose'] == 'For Sale')].index, inplace=True)

for distinct_items in df['property_type'].unique():
    print_metadata(distinct_items)

drop_irrelevant_price_rows(None, 300000, 600000000, 600000000, 'Flat')
drop_irrelevant_price_rows(21000, 100000, 600000000, 600000000, 'House')
drop_irrelevant_price_rows(None, 200000, 600000000, 600000000, 'Farm House')
drop_irrelevant_price_rows(None, 50000, 600000000, 600000000, 'Lower Portion')
drop_irrelevant_price_rows(None, 50000, 600000000, 600000000, 'Upper Portion')
drop_irrelevant_price_rows(None, 3000, 600000000, 600000000, 'Room')
print(df.info)
```

```python
# converting categorical into numerical columns finding median of each mean and putting them accordingly
def fix_categorical_values(column_name):
    distinct_categorical_values = df[column_name].unique()
    
    # finding mean and assigning into actual dataframe
    for value in distinct_categorical_values:
        median = df.loc[(df[column_name] == value), 'price'].median()
        df.loc[(df[column_name] == value), column_name] = median
    
fix_categorical_values('province_name')
fix_categorical_values('city')
fix_categorical_values('property_type')
fix_categorical_values('location')

# for purpose property_type widely effect mean prices so we do analysis according to it
def fix_categorical_purpose_value():
    purpose_values_distinct = df['purpose'].unique()
    property_types_distinct = df['property_type'].unique()
    
    for purpose in purpose_values_distinct:
        for property_type in property_types_distinct:
            median = df.loc[((df['purpose'] == purpose) & (df['property_type'] == property_type)), 'price'].median()
            df.loc[((df['purpose'] == purpose) & (df['property_type'] == property_type)), 'purpose'] = median

fix_categorical_purpose_value()

print(df['purpose'].unique())
print(df['province_name'].unique())
print(df['city'].unique())
print(df['property_type'].unique())
print(df['location'].unique())
```

## Regression Model Started

```python
from sklearn.model_selection import train_test_split
# df2 = df # backup

# X = df.drop(['province_name', 'city', 'property_type', 'purpose'], axis=1)
X = df.drop('price', axis=1)
y = df['price']

print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
print('X_train', X_train)
print('X_test', X_test)
print('y_train', y_train)
print('y_test', y_test)
```

```python
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

regressor = linear_model.LinearRegression()

regressor.fit(X_train_scaled, y_train)
```

```python
y_pred = regressor.predict(X_test_scaled)
print('predicted y', y_pred)
print('actual y', y_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_val_score

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
cross_val_scores = cross_val_score(regressor, X, y, cv=5, scoring='r2')

print('r squared', r2)
print('mean absolute error', mae)
print('mean squared error', mse)
print('root mean squared error', rmse)
print('cross validation score', cross_val_scores)
```
